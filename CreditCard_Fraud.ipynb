```python
from sklearn.model_selection import train_test_split
```

```python
Y = credit_df['Class']
X= credit_df.drop(['Class'],axis=1)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.70, test_size=0.30, random_state=1)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix,mean_squared_error,balanced_accuracy_score
```

```python
model = IsolationForest(n_estimators=5)
model.fit(X_train)
```

```python
#get the anomaly score
anomalyS_train = model.decision_function(X_train)
anomaly_pred = model.predict(X_train)
```
